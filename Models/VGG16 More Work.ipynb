```python
from keras.models import Model
from keras.models import Sequential
from keras.layers import Dense, Flatten, Dropout
from keras.optimizers import Adam
from keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import pandas as pd
import tensorflow as tf
from keras.applications.vgg16 import VGG16
```

```python
train_data = np.load('drive/MyDrive/Deep Learning Project/Data/CIFAKE_Train.npz')
test_data = np.load('drive/MyDrive/Deep Learning Project/Data/CIFAKE_Test.npz')
```

```python
x_train = train_data['images']
y_train = train_data['labels']
```

```python
y_train = y_train.astype('int')
```

```python
x_test = test_data['images']
y_test = test_data['labels']
```

```python
m_train = x_train.shape[0]
x_train = x_train.reshape(m_train, 32 * 32 * 3)

m_test = x_test.shape[0]
x_test = x_test.reshape(m_test, 32 * 32 * 3)

x_train.shape
```

```python
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)

x_test = scaler.transform(x_test)

x_train.min(), x_train.max()
```

```python
x_train = x_train.reshape(m_train, 32, 32, 3)
x_test = x_test.reshape(m_test, 32, 32, 3)
x_train.shape
```

```python
vgg16 = VGG16(weights='imagenet', include_top=False, input_shape=(32,32,3))
```

```python
model = Sequential()
model.add(vgg16)
model.add(Flatten())
model.add(Dense(256,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
model.layers[0].trainable=False
```

```python
model.compile(loss='binary_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
epochs = 20
batch_size = 64
model.fit(x_train, y_train, epochs=epochs, batch_size=batch_size, validation_split=0.2)
```

```python
test_loss, test_accuracy = model.evaluate(x_test, y_test)
```
